```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/sunnysavita10/credit_card_pw_hindi/main/creditCardFraud_28011964_120214.csv")
```

```python
x = df.drop('default payment next month',axis=1)
```

```python
y = df['default payment next month']
```

```python
x
```

```python
y
```

```python
pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=69)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train.max()
```

```python
x_train.min()
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
```

```python
nb.fit(x_train,y_train)
```

```python
y_predicted = nb.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print(accuracy_score(y_test,y_predicted)*100)
print(confusion_matrix(y_test,y_predicted))
print(classification_report(y_test,y_predicted))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.naive_bayes import GaussianNB
new_nb = GaussianNB()
```

```python
grid = {'var_smoothing' : [1e-9,0.1, 0.001, 0.5,0.05,0.01,1e-8,1e-7,1e-6,1e-10,1e-11]}
```

```python
gs = GridSearchCV(new_nb,param_grid=grid,cv=5,verbose=3)
```

```python
gs.fit(x_train,y_train)
```

```python
gs.best_params_
```

```python
y_predicted_new = gs.predict(x_test)
```

```python
print(accuracy_score(y_test,y_predicted_new)*100)
print(confusion_matrix(y_test,y_predicted_new))
print(classification_report(y_test,y_predicted_new))
```

```python
from sklearn.svm import SVC
svc = SVC()
```

```python
grid = {'C':[0.1,0.5,1,10,100],
        'kernel':['linear', 'poly', 'rbf', 'sigmoid'],
        'gamma':['scale','auto']}
```

```python
svc_gs = GridSearchCV(svc,param_grid=grid,verbose=3)
```

```python
svc_gs.fit(x_train,y_train)
```

```python
svc_gs.best_params_
```

```python
y_predicted_svc = svc_gs.predict(x_test)
```

```python
print(accuracy_score(y_test,y_predicted_svc)*100)
print(confusion_matrix(y_test,y_predicted_svc))
print(classification_report(y_test,y_predicted_svc))
```

```python
from sklearn.naive_bayes import GaussianNB
opt_NB = GaussianNB(var_smoothing = 0.5)
opt_NB.fit(x_train,y_train)
y_predicted_optNB = opt_NB.predict(x_test)
```

```python
print(accuracy_score(y_test,y_predicted_optNB)*100)
print(confusion_matrix(y_test,y_predicted_optNB))
print(classification_report(y_test,y_predicted_optNB))
```
